# Using financialdatapy package

```python
from financialdatapy.stock import Stock
```

## US Companies

```python
apple = Stock('aapl')
```

### Financial Statements as Reported

#### Annual

**Income Statement**

```python
income_statement = apple.financials('income_statement', 'annual')
income_statement
```

**Balance Sheets**

```python
balance_sheet = apple.financials('balance_sheet')  # period parameter defaults to 'annual'
balance_sheet
```

**Cash Flow**

```python
cash_flow = apple.financials('cash_flow')
cash_flow
```

#### Quarter

**Income Statement**

```python
income_statement_q = apple.financials('income_statement', 'quarter')
income_statement_q
```

**Balance Sheets**

```python
balance_sheet_q = apple.financials('balance_sheet', 'quarter')
balance_sheet_q
```

**Cash Flow**

```python
cash_flow_q = apple.financials('cash_flow', 'quarter')
cash_flow_q
```

### Standard Financial Statements

#### Annual

**Income Statement**

```python
std_ic = apple.financials('income_statement', 'annual', is_standard=True)
std_ic
```

**Balance Sheets**

```python
std_bs = apple.financials('balance_sheet', is_standard=True)
std_bs
```

**Cash Flow**

```python
std_cf = apple.financials('cash_flow', is_standard=True)
std_cf
```

#### Quarterly

**Income Statement**

```python
std_ic_q = apple.financials('income_statement', 'quarter', is_standard=True)
std_ic_q
```

**Balance Sheets**

```python
std_bs_q = apple.financials('balance_sheet', 'quarter', is_standard=True)
std_bs_q
```

**Cash Flow**

```python
std_cf_q = apple.financials('cash_flow', 'quarter', is_standard=True)
std_cf_q
```

### Historical Price

```python
apple_price = apple.price('2021-1-1', '2021-1-31')
apple_price
```
